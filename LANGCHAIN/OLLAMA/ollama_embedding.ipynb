```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(
    OllamaEmbeddings(model="embeddinggemma:latest")  ##by default it uses llama 2
)
```

```python
r1=embeddings.embed_documents(
    [
        "aplha is the first letter",
        "beta is the second letter"
    ]
)
```
